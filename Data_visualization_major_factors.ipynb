```python
import numpy as np
import pandas as pd
from plotnine import *

%matplotlib inline
df = pd.read_csv('boko_kill3.csv').dropna() #removes NAs!
```

```python
ggplot(df, aes(x='imonth', y='nwound', color = 'nkill')) + geom_point() + geom_smooth() + labs(y = "Numbers wounded", x ="Month") + ggtitle("Relationship between month and numbers wounded by Boko Haram") 
```

```python
# normal distribution 

ggplot(df, aes(x='iday', y='nwound', color = 'nkill')) + geom_point() + geom_smooth() #adding regression line
```

```python
ggplot(df, aes(x='imonth', y='iday')) + geom_point() + geom_smooth() #adding regression line
#normal distribution
```

```python
ggplot(df, aes(x='nwound', y='nkill')) + geom_point() + geom_smooth()+ labs(x = "Numbers wounded", y ="Numbers Killed") + ggtitle("Relationship between numbers wounded and killed by Boko Haram") 
```

```python
ggplot(df, aes(x='nkill', y='nwound')) + geom_point() + geom_smooth() #adding regression line
```

```python
ggplot(df, aes(x='nkill')) + \
    geom_bar(stat = 'count')
```

```python
ggplot(df, aes(x='nwound')) + \
    geom_bar(stat = 'count')

```

```python
## normal distribution

ggplot(df, aes(x='imonth', y='nwound', color = 'nkill'))+ geom_boxplot()  + \
geom_smooth() #adding regression line
```

```python
#month 6.2 most wounding occurs - about Jun/weeks 2-3
#normal distribution of wounding 

ggplot(df, aes(x='imonth', y='nwound'))+ \
    geom_boxplot() 
```
