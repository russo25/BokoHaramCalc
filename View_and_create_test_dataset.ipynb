# Reading Boko Haram (BH) CSV File Into a Dictionary With csv
#STEP 1-Read csv file, view shape, create test dataset of 20% (test_boko.csv)

You can read CSV data directly into a dictionary.

See https://realpython.com/python-csv/

```python
# Here’s the code to read "boko_kill.csv" as a dictionary 
import csv

with open('boko_kill.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            print(f'\n\timonth\tiday\tnwound\tnkill')
            line_count += 1
        else:
            print(f'\t{row["imonth"]}\t\t{row["iday"]}\t\t{row["nwound"]}\t\t{row["nkill"]}.')
            line_count += 1
    print(f'Processed {line_count} lines.')
```

```python
# Reading the CSV into a panda DataFrame
import pandas
df = pandas.read_csv('boko_kill.csv')

print(df.size)
print(df.shape)
print(df)
```

```python
# Using Pandas, select  the 4 columns of interest
fields = ["imonth", "iday", "nkill", "nwound"]

print(df[fields])
```

```python
# Read only the desired fields into memory
df = pandas.read_csv('boko_kill.csv', skipinitialspace=True, usecols=fields)

print(df.size)
print(df.shape)
print(df)
```

```python
# Print first row's data
df.iloc[0]
```

```python
# Print first row's data as a row
print(df.iloc[0:1])
df.iloc[0:1]
```

```python
# Print first 3 rows of data as a row
print(df.iloc[0:3])
df.iloc[0:3]
```
